## dilling with missing values

i'm going to fill the missing values in the year column, by finding the mean of year for every platform, by using the groupby and transform method. 
i did that, because there are not many missing values in the year column, and u can pretty much determine the year of the game based on the platform. thats because allot of games are made for specific platforms, that where sold in a specific period of time. 

```python
games_df['year_of_release'] = games_df.groupby(['platform'])['year_of_release'].transform(lambda grp: grp.fillna(np.mean(grp)))
games_df.info()
```

# analyze the data

## year significance

by using box plot and a histogram, i'll see sales based on years. and i'll also see if there is to much of a skew in the data. 

```python
games_df["year_of_release"].value_counts()
```

```python
def boxplot_games(column):
    Q1 = games_df[column].quantile(0.25)
    Q3 = games_df[column].quantile(0.75)
    IQR = Q3 - Q1
    games_df.boxplot(column)
    plt.hlines(y = Q1-1.5*IQR, xmin = 0.9, xmax = 1.1, color = "red")
    plt.hlines(y = Q3 + 1.5*IQR, xmin = 0.9, xmax = 1.1, color = "red")
```

```python
games_df.plot(y = "year_of_release", kind = "hist", bins = 37, xlim = (1975,2020), ylim = (0,1750))
plt.title('year_production')
plt.xlabel('year')
plt.ylabel('num_of_games')
plt.show()
```

```python
plt.ylim(1970,2025)
boxplot_games("year_of_release")
plt.title('production_range')
plt.ylabel('year')
plt.show()
```

seems that the years with the most sales are around 2005- 2010. their weren't a lot of games before the 90's bot they are not harming the data.

# test hypotheses

now using hypotheses on equality of two population means, im going to check tow hypothesis.
1. h1: the average user score of xone is equal to pc.
1. h2: the average user score of action is different from sports.

lets start with h1. 

## xbox vs pc

```python
xbox1_df=relevant_years_df[(relevant_years_df['platform']=='XOne')& (relevant_years_df['user_score'].notnull())]
pc_df=relevant_years_df[(relevant_years_df['platform']=='PC')& (relevant_years_df['user_score'].notnull())]
print(np.var(xbox1_df["user_score"]))
np.var(pc_df["user_score"])
```

```python
print(xbox1_df["user_score"].mean())
pc_df["user_score"].mean()
```

```python
alpha = 0.05

results = st.ttest_ind(xbox1_df["user_score"], pc_df["user_score"],equal_var=False)

print('p-value: ', results.pvalue)

if results.pvalue < alpha:
    print("We reject the null hypothesis")
else:
    print("We can't reject the null hypothesis") 
```
